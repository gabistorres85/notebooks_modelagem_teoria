## SOMA DO QUADRADOS DOS ERROS - SQE

$$
SQE = \sum {(y_i-\hat{y_1}})^2
$$

## Estimativa da variância $\sigma^2$ dos erros


$$
\hat{\sigma^2}=s^=\frac{SQE}{n-2}
$$

## Soma dos Quadrados Totais

$$
SQT = \sum{(y - \overline{y})^2}
$$

## Coeficiente de Determinação

$$
r^2 = 1 - \frac{SQE}{SQT}
$$

```python
# Soma dos quadrados dos erros - estimativa da variância
sqe = np.sum(np.square(residuo))
variancia = sqe/(n-2)
dp = np.sqrt(variancia)
print(f'Resultado do SQE = {sqe:.4f}\nVariancia estimada = {variancia:.4f}\ndesvio padrao = {dp:.3f}')
```
